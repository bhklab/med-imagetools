```python
# ruff: noqa
from pathlib import Path

from rich import print as rprint

from imgtools.dicom.crawl import CrawlerSettings
from imgtools.io.loaders import ImageMaskInput, ImageMaskModalities
from imgtools.io.writers import ImageMaskOutput
```

```python
datadir = Path('testdata/Head-Neck-PET-CT')
crawler_settings = CrawlerSettings(
    dicom_dir=datadir,
    n_jobs=4,
)
loader = ImageMaskInput(crawler_settings=crawler_settings, modalities=ImageMaskModalities.CT_RTSTRUCT)
```

```python
first = loader.keys()[0]
first
```

```python
data = loader(first)
```

```python
print(f'{data.scan!r}')
```

```python
print(f'{data.mask!r}')
```

```python
# context_keys is not predictable in the current implementation
# temporary workaround until metadata is standardized
context_keys = list(data.scan.metadata.keys()) + list(data.mask.metadata.keys())


# instantiate
output = ImageMaskOutput(
    context_keys, 
    root_directory=datadir.parent / "niftis", 
    filename_format="{CaseID}/{Modality}_Series-{SeriesInstanceUID}/{ImageID}.nii.gz"
)
print(f'{output!r}')
print(f'{output.writer.pattern_resolver!r}')
```

```python
output(data)
```

```python
import pandas as pd
pd.read_csv(output.writer.index_file).dropna(axis=1, how='all')
```

```python
for key in loader.keys()[:3]:
    data = loader(key)
    output(data)

pd.read_csv(output.writer.index_file).dropna(axis=1, how='all')
```
