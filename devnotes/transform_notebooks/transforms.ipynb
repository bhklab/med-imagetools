```python
import SimpleITK as sitk
from imgtools.datasets.sample_images import (
    create_sphere_image,
    create_grid_image,
    create_gradient_image,
    create_cross_image,
    create_rod_image,
    create_noisy_sphere_image,
    create_checkerboard_image,
    create_ct_hounsfield_image,
)

from imgtools.vizualize import (
    ImageVisualizer,
    display_slices,
)
```

## Sphere

```python
sphere = create_sphere_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    radius=20,
)

viz = ImageVisualizer.from_image(sphere)
viz.view_slices()
```

## Grid

```python
grid = create_grid_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    grid_spacing=8,
)

viz = ImageVisualizer.from_image(grid)
viz.view_slices()
```

## Gradient

```python
# X-axis gradient
x_gradient = create_gradient_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    direction="x",
    min_value=0.0,
    max_value=1.0,
)

viz = ImageVisualizer.from_image(x_gradient)
viz.view_slices()
```

```python
# Radial gradient
radial_gradient = create_gradient_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    direction="radial",
    min_value=0.0,
    max_value=1.0,
)

viz = ImageVisualizer.from_image(radial_gradient)
viz.view_slices()
```

## Cross

```python
cross = create_cross_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    thickness=20,
)

viz = ImageVisualizer.from_image(cross)
viz.view_slices()
```

## Rod

```python
# X-axis rod
x_rod = create_rod_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    axis="x",
    radius=5,
)

viz = ImageVisualizer.from_image(x_rod)
viz.view_slices()
```

```python
# Z-axis rod
z_rod = create_rod_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    axis="z",
    radius=5,
)

viz = ImageVisualizer.from_image(z_rod)
viz.view_slices()
```

## Noisy Sphere

```python
noisy_sphere = create_noisy_sphere_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
    radius=20,
    noise_level=0.3,
)

viz = ImageVisualizer.from_image(noisy_sphere)
viz.view_slices()
```

## Checkerboard

```python
checkerboard = create_checkerboard_image(
    # size=(100, 100, 100),
    # spacing=(1.0, 1.0, 1.0),
    # checker_size=10,
)

viz = ImageVisualizer.from_image(checkerboard)
viz.view_slices()
```

## CT Hounsfield Unit Image

```python
ct_image = create_ct_hounsfield_image(
    size=(100, 100, 100),
    spacing=(1.0, 1.0, 1.0),
)

viz = ImageVisualizer.from_image(ct_image)
viz.view_slices()
```

```python
# Calculate statistics on the CT image
stats = sitk.StatisticsImageFilter()
stats.Execute(ct_image)
print(f"CT image value range: [{stats.GetMinimum():.1f}, {stats.GetMaximum():.1f}]")
print(f"Mean value: {stats.GetMean():.1f}")
print(f"Standard deviation: {stats.GetSigma():.1f}")
```

## Combined Visualization

```python
# Create a visualization with multiple images to compare
images = {
    "Sphere": sphere,
    "Grid": grid,
    "X-Gradient": x_gradient,
    "Radial": radial_gradient,
    "Cross": cross,
    "Rod-X": x_rod,
    "Rod-Z": z_rod,
    # "Noisy Sphere": noisy_sphere,
    "Checkerboard": checkerboard,
    "CT Image": ct_image
}

# Display middle slice of each image
for name, image in images.items():
    print(f"\n{name}:")
    size = image.GetSize()

    middle_slice = size[2] // 2
    img = sitk.GetArrayViewFromImage(image)
    display_slices(img, x_index=middle_slice, y_index=middle_slice, z_index=middle_slice)
```
